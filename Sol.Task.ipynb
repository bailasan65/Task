```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('imdb_top_1000.csv')
data.describe
```

```python
edited_data = data.drop(columns=['Poster_Link', 'Released_Year', 'Certificate', 'Runtime', 'Overview', 'Meta_score', 'Star1', 
                      'Star2', 'Star3', 'Star4', 'Gross'])
edited_data.fillna(edited_data.mean())
```

```python
genre_counts = data['Genre'].value_counts()


genre_counts.idxmax()
```

```python
countdata=pd.DataFrame({'Genre': genre_counts.index, 'Count': genre_counts.values})
print(countdata)
```

```python
countdata.plot.bar(x='Genre', y='Count', figsize=(40,20),legend=False)

plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('number of Movies,TV Shows by Genree')
plt.xticks(rotation=90)
```

```python
dirc = data['Director'].value_counts().reset_index()

# Rename the columns
dirc.columns = ['Director', 'Count']

# Sort the values by count in descending order
dirc = dirc.sort_values(by='Count', ascending=False)
```

```python
plt.figure(figsize=(40,20))


plt.bar(dirc['Director'], dirc['Count'])

plt.xlabel('Director')
plt.ylabel('Count')
plt.title('Director  Most Top rated Movies||TV Shows')

plt.xticks(rotation=90)

plt.show()
```
